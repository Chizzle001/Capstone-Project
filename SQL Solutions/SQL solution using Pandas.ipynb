# `Pandas SQL`

```python
import pandas as pd

movies_df = pd.read_csv('Movies_file_1.csv')
stars_df = pd.read_csv('Movies_file_2.csv')
```

```python
# Table 1: movies_df
# 1) Display all the details of movies created by directors Christopher and Matt Reeves.
result1 = movies_df[(movies_df['Director1'].isin(['Christopher Nolan', 'Matt Reeves'])) |
                    (movies_df['Director2'].isin(['Christopher Nolan', 'Matt Reeves'])) |
                    (movies_df['Director3'].isin(['Christopher Nolan', 'Matt Reeves']))]
print("Movies created by directors Christopher and Matt Reeves:")
result1
```

```python
movies_df['Duration'] = movies_df['Duration'].str.replace('min', '')  # Remove 'min'
movies_df['Duration'] = movies_df['Duration'].astype(int)  # Convert to integer
```

```python
# 2) Display all the details of movies with a duration of 140 minutes to 190 minutes.
movies_df['Duration'] = movies_df['Duration'].astype(int)  # Convert 'Duration' column to integer
result2 = movies_df[(movies_df['Duration'] >= 140) & (movies_df['Duration'] <= 190)]
print("Movies with a duration of 140 minutes to 190 minutes:")
result2
```

```python
# 3) Display all details of movies with ratings above 7 in ascending order.
result3 = movies_df[movies_df['Ratings'] > 7].sort_values('Ratings')
print("Movies with ratings above 7 in ascending order:")
result3
```

```python
# 4) Display all movie names in descending order.
result4 = movies_df.sort_values('Movie_Name', ascending=False)[['Movie_Name']]
print("Movie names in descending order:")
result4
```

```python
# 5) Display movie names starting with 'P' and their rating is greater than 7.
result5 = movies_df[(movies_df['Movie_Name'].str.startswith('P')) & (movies_df['Ratings'] > 7)]
print("Movie names starting with 'P' and their rating is greater than 7:")
result5
```

# `Table 2`

```python
# Query 1: Display all movie names with star Arnold Schwarzenegger in ascending order
result1 = stars_df[stars_df['Star1'] == 'Arnold Schwarzenegger']['Movie_Name'].sort_values(ascending=True)
print("Movie names with star Arnold Schwarzenegger in ascending order:")
result1
```

```python
# Query 2: Display all details of the movie with the highest number of votes
max_votes = stars_df['Votes'].max()
result2 = stars_df[stars_df['Votes'] == max_votes]
print("Details of the movie with the highest number of votes:")
result2
```

```python
# Query 3: Display movie names with gross collections in descending order
result3 = stars_df.sort_values('Gross_Collection', ascending=False)['Movie_Name']
print("Movie names with gross collections in descending order:")
result3
```

```python
# Query 4: Display the gross collection of movies with the star Arnold
result4 = stars_df[stars_df[['Star1', 'Star2', 'Star3', 'Star4']].apply(lambda row: 'Arnold Schwarzenegger' in row.values, axis=1)]['Gross_Collection']
print("Gross collection of movies with the star Arnold:")
result4
```

```python
# Query 5: Display all details of movies with comedy and action genres
result5 = stars_df[(stars_df['Genre1'] == 'Action') & (stars_df['Genre2'] == 'Comedy')]
print("Details of movies with comedy and action genres:")
result5
```

# `Table 3`

```python
# Query 1: Display all details from both tables where movie names are the same
result1 = pd.merge(movies_df, stars_df, on='Movie_Name')
print("Details from both tables where movie names are the same:")
result1
```

```python
# Query 3: Display all details from both tables with the highest gross collection
max_gross_collection = max(stars_df['Gross_Collection'])
result3 = movies_df.merge(stars_df[stars_df['Gross_Collection'] == max_gross_collection], on='Movie_Name')
print("Details from both tables with the highest gross collection:")
result3
```

```python
# Get the movie with the highest ratings
max_ratings = movies_df['Ratings'].max()

# Perform inner join on movies_df and stars_df based on Movie_Name and filter by highest ratings
result4 = movies_df.merge(stars_df, on='Movie_Name')
result4 = result4[result4['Ratings'] == max_ratings]

print("Details from both tables with the highest ratings:")
result4
```

```python
min_gross_collection = min(stars_df['Gross_Collection'])
min_ratings = min(movies_df['Ratings'])
result5 = movies_df.merge(stars_df[(stars_df['Gross_Collection'] == min_gross_collection) & (movies_df['Ratings'] == min_ratings)], on='Movie_Name')
print("Details from both tables with the lowest gross collection and lowest ratings:")
result5
```
