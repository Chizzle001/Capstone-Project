```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('movies.db')
cursor = conn.cursor()

# Create the first table
cursor.execute('''CREATE TABLE IF NOT EXISTS movies (
                Movie_Name TEXT,
                Director1 TEXT,
                Director2 TEXT,
                Director3 TEXT,
                Duration TEXT,
                Year INTEGER,
                Ratings REAL,
                Metascore INTEGER,
                Sno INTEGER
                )''')

# Read the first CSV file and insert the data into the 'movies' table
df1 = pd.read_csv('Movies_file_1.csv')
df1.to_sql('movies', conn, if_exists='replace', index=False)

# Create the second table
cursor.execute('''CREATE TABLE IF NOT EXISTS stars (
                Movie_Name TEXT,
                Star1 TEXT,
                Star2 TEXT,
                Star3 TEXT,
                Star4 TEXT,
                Votes INTEGER,
                Genre1 TEXT,
                Genre2 TEXT,
                Genre3 TEXT,
                Gross_Collection TEXT,
                Popularity REAL,
                Certification TEXT
                )''')

# Read the second CSV file and insert the data into the 'stars' table
df2 = pd.read_csv('Movies_file_2.csv')
df2.to_sql('stars', conn, if_exists='replace', index=False)

# Commit the changes and close the connection
conn.commit()
conn.close()
```

```python
# Query 1: Display all the details of movies created by directors Christopher Nolan and Matt Reeves
query1 = '''SELECT * FROM movies
            WHERE Director1 = 'Christopher Nolan'
                OR Director1 = 'Matt Reeves'
                OR Director2 = 'Christopher Nolan'
                OR Director2 = 'Matt Reeves'
                OR Director3 = 'Christopher Nolan'
                OR Director3 = 'Matt Reeves' '''
df_query1 = pd.read_sql_query(query1, conn)
print("Details of movies created by directors Christopher Nolan and Matt Reeves:")
df_query1
```

```python
# Query 2: Display all the details of movies with a duration of 140 minutes to 190 minutes
query2 = '''SELECT * FROM movies
            WHERE Duration >= '140 min' AND Duration <= '190 min' '''
df_query2 = pd.read_sql_query(query2, conn)
print("Details of movies with a duration of 140 minutes to 190 minutes:")
df_query2
```

```python
# Query 3: Display all details of movies with ratings above 7 in ascending order
query3 = '''SELECT * FROM movies
            WHERE Ratings > 7
            ORDER BY Ratings ASC '''
df_query3 = pd.read_sql_query(query3, conn)
print("Details of movies with ratings above 7 in ascending order:")
df_query3
```

```python
# Query 4: Display all movie names in descending order
query4 = '''SELECT Movie_Name FROM movies
            ORDER BY Movie_Name DESC '''
df_query4 = pd.read_sql_query(query4, conn)
print("Movie names in descending order:")
df_query4
```

```python
# Query 5: Display movie names starting with 'P' and their rating is greater than 7
query5 = '''SELECT Movie_Name, Ratings FROM movies
            WHERE Movie_Name LIKE 'P%' AND Ratings > 7 '''
df_query5 = pd.read_sql_query(query5, conn)
print("Movie names starting with 'P' and their rating is greater than 7:")
df_query5
```

## `Make SubQueries`

```python
# Query 1: Display all details from both tables where movie names are the same
query1 = '''SELECT * FROM movies
            INNER JOIN stars ON movies.Movie_Name = stars.Movie_Name'''
df_query1 = pd.read_sql_query(query1, conn)
print("Details from both tables where movie names are the same:")
df_query1
```

```python
# Query 2: Display all movie names, Director, ratings, and gross collection where the genre is action
query2 = '''SELECT movies.Movie_Name, movies.Director1, movies.Ratings, stars.Gross_Collection 
            FROM movies
            INNER JOIN stars ON movies.Movie_Name = stars.Movie_Name
            WHERE stars.Genre1 = 'Action' OR stars.Genre2 = 'Action' OR stars.Genre3 = 'Action' '''
df_query2 = pd.read_sql_query(query2, conn)
print("Movie names, Director, ratings, and gross collection where the genre is action:")
df_query2
```

```python
# Query 3: Display all details from both tables with the highest gross collection
query3 = '''SELECT * FROM movies
            INNER JOIN stars ON movies.Movie_Name = stars.Movie_Name
            WHERE stars.Gross_Collection = (SELECT MAX(Gross_Collection) FROM stars)'''
df_query3 = pd.read_sql_query(query3, conn)
print("Details from both tables with the highest gross collection:")
df_query3
```

```python
# Query 4: Display all details from both tables with the highest ratings
query4 = '''SELECT * FROM movies
            INNER JOIN stars ON movies.Movie_Name = stars.Movie_Name
            WHERE movies.Ratings = (SELECT MAX(Ratings) FROM movies)'''
df_query4 = pd.read_sql_query(query4, conn)
print("Details from both tables with the highest ratings:")
df_query4
```

```python
# Query 5: Display the movie with the lowest rating and gross collection
query5 = '''SELECT m.Movie_Name, m.Director1, m.Ratings, s.Gross_Collection
            FROM movies AS m
            INNER JOIN stars AS s ON m.Movie_Name = s.Movie_Name
            WHERE m.Ratings = (SELECT MIN(Ratings) FROM movies)
            AND s.Gross_Collection = (SELECT MIN(Gross_Collection) FROM stars)'''
df_query5 = pd.read_sql_query(query5, conn)
print("Movie with the lowest rating and gross collection:")
df_query5
```
